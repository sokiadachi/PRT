```python
#numpy,scipy.statsからnorm,math,matplotlib.pyplotをインポート
import numpy as np
from scipy.stats import norm
import math
import matplotlib.pyplot as plt

#0から1まで、0.01間隔で入ったリストXb(改良前用)とXa(改良後用)を作成
Xb0 = np.arange(0,1,0.01)
Xb1 = np.arange(0,1,0.01)
Xb2 = np.arange(0,1,0.01)
Xb3 = np.arange(0,1,0.01)
Xb4 = np.arange(0,1,0.01)
Xb5 = np.arange(0,1,0.01)

#回数ごとの実験結果
a1 = [1]
a2 = [1, 0]
a3 = [1, 0, 0]
a4 = [1, 0, 0, 1]
a5 = [1, 0, 0, 1, 1]

#標準偏差を求める
sd1 = np.std(a1)
sd2 = np.std(a2)
sd3 = np.std(a3)
sd4 = np.std(a4)
sd5 = np.std(a5)

print(sd1, sd2, sd3, sd4, sd5)

#分散を求める
var1 = np.var(a1)
var2 = np.var(a2)
var3 = np.var(a3)
var4 = np.var(a4)
var5 = np.var(a5)

print(var1, var2, var3, var4, var5)

#回数ごとの標準偏差
b1 = [sd1]
b2 = [sd1, sd2]
b3 = [sd1, sd2, sd3]
b4 = [sd1, sd2, sd3, sd4]
b5 = [sd1, sd2, sd3, sd4, sd5]

#回数ごとの標準偏差の標準偏差
sd11 = np.std(b1)
sd12 = np.std(b2)
sd13 = np.std(b3)
sd14 = np.std(b4)
sd15 = np.std(b5)

print(sd11, sd12, sd13, sd14, sd15)

#確率密度関数にXb(改良前),平均、標準偏差を代入
Yb0 = norm.pdf(Xb0,0.5,100)
Yb1 = norm.pdf(Xb1,1,100)
Yb2 = norm.pdf(Xb2,0.5,sd12)
Yb3 = norm.pdf(Xb3,1/3,sd13)
Yb4 = norm.pdf(Xb4,0.5,sd14)
Yb5 = norm.pdf(Xb5,0.6,sd15)

#x,yを引数にして、関数の色をr(red)に指定
#plt.plot(Xb0,Yb0,color='r')
plt.plot(Xb1,Yb1,color='g')
plt.plot(Xb2,Yb2,color='b')
plt.plot(Xb3,Yb3,color='c')
plt.plot(Xb4,Yb4,color='m')
plt.plot(Xb5,Yb5,color='y')

plt.show()
```
